```python
df.head()
```

```python
plt.plot(df["index"],df["V Kohli"])
```

```python
plt.plot(df["index"],df["RG Sharma"])
```

```python
plt.plot(df["index"],df["V Kohli"])
plt.plot(df["index"],df["RG Sharma"])
```

```python
plt.plot(df["index"],df["V Kohli"],linestyle="dotted")
plt.plot(df["index"],df["RG Sharma"],linestyle="dashed")
plt.title("RG VS VK")
plt.xlabel("Season")
plt.ylabel("Runs")
```

```python
plt.plot(df["index"],df["V Kohli"],linestyle="dotted",color="black")
plt.plot(df["index"],df["RG Sharma"],linestyle="dashed",color="indigo")
plt.title("RG VS VK")
plt.xlabel("Season")
plt.ylabel("Runs")
```

```python
plt.plot(df["index"],df["V Kohli"],linestyle="dotted",color="black",linewidth="2")
plt.plot(df["index"],df["RG Sharma"],linestyle="dashed",color="indigo",linewidth="2")
plt.title("RG VS VK")
plt.xlabel("Season")
plt.ylabel("Runs")
```

```python
plt.plot(df["index"],df["V Kohli"],linewidth=2,color="black",linestyle="dotted",marker="d") # marker=shapes(s,o,h,d,p,v,x,+,1,)
plt.plot(df["index"],df["RG Sharma"],linewidth=2,linestyle="dashed",color="indigo",marker="*")
plt.title("RG VS VK")
plt.xlabel("Season")
plt.ylabel("Runs")
plt.grid()
```

```python
df2=pd.read_csv("batsman_season_record.csv")
df2
```

```python
plt.bar(df2["batsman"],df2["2016"])
```

```python
plt.bar(df2["batsman"],df2["2017"],label='2017',color='red')
plt.bar(df2["batsman"],df2["2016"],bottom=df2["2017"],label="2016",color='black')
plt.bar(df2["batsman"],df2["2015"],bottom=df2["2016"]+df2['2017'],label="2015",color='grey')
plt.grid()
```

```python
#histogram - univarient analysis
#            used for numeric column
#            used to count the frequency
data=[34,55,66,44,76,56,48,31,45]
plt.hist(data)
```

```python
#pie chart = used for univarient analysis and bi varient analysis
#            categorical vs numerical data
data2=['45','55','78','65']
sub=['eng','phy','math','chem']
plt.pie(data2,labels=sub)
plt.show()
```
