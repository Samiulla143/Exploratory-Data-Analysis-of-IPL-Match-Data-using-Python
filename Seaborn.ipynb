```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import plotly.express as px # inbuilt data
```

```python
df=sns.load_dataset('tips')
df.head(5)
```

```python
#2 ways to call scatter plot
#1 relplot
#2 scatterplot
```

```python
sns.relplot(x='total_bill',y='tip',data=df)
```

```python
#hue
sns.relplot(x='total_bill',y='tip',hue='smoker',data=df)
```

```python

sns.catplot(x='day',y='tip',kind='strip',data=df)
```

```python

sns.catplot(x='day',y='tip',kind='swarm',data=df)
```

```python

sns.catplot(x='day',y='tip',kind='strip',jitter=0,data=df)
```

```python
#line plot
gap=px.data.gapminder()
gap.head()
```

```python
sns.lineplot(data=gap,x='year',y='lifeExp')
```

```python
sns.relplot(data=gap,x='year',y='lifeExp',kind='line')
```

```python
temp_df=gap[gap['country'].isin(['India','Pakistan','China'])]
temp_df
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country')
plt.show()
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country',style='continent')
plt.show()
```

```python
#Fect plot - represnt same plot in diff axis
sns.relplot(kind='scatter',data=df,x='total_bill',y='tip',col='sex')
plt.show()
```

```python
#distribution plot=univarient analysis
                #used to find distribution
#central tendency
sns.histplot(data=df,x='total_bill')
```

```python
sns.histplot(data=df,x='total_bill',bins=20,hue='sex')
```

```python
#KDE plot-kernel density plot
sns.kdeplot(data=df,x='total_bill')
```

```python
#heat map -co relation
t=gap.pivot(index='country',columns='year',values='lifeExp')
sns.heatmap(t)
```

```python
t=gap[gap['continent']=='Americas'].pivot(index='country',columns='year',values='lifeExp')
sns.heatmap(t)
```

```python
sns.boxplot(data=df,x='day',y='total_bill')
plt.show()
```

```python
#vilon plot
sns.violinplot(data=df,x='day',y='total_bill')
plt.show()
```

```python
sns.barplot(data=df,x='day',y='total_bill',hue='smoker')
plt.show()
```

```python
#countplot- a special case of barplot in which we can show the number of oservation in each category
sns.countplot(data=df,x='sex',hue='day')
```
